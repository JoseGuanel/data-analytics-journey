## 🔢 NumPy

### Crear arrays
- Array 1D: np.array([1, 2, 3])
- Array 2D: np.array([[1,2],[3,4]])
- Rango: np.arange(inicio, fin, paso)
- Aleatorio enteros: np.random.randint(min, max, size=n)
- Semilla fija: np.random.seed(42)

### Indexing y Slicing
- Elemento: array[i]
- Rango: array[inicio:fin]
- Saltar: array[::paso]
- 2D: array[fila, columna]
- Toda fila: array[fila, :]
- Toda columna: array[:, columna]

### Operaciones
- Suma: np.sum(array, axis=0/1)
- Promedio: np.mean(array, axis=0/1)
- Máximo: np.max(array) | Índice: np.argmax(array)
- Mínimo: np.min(array) | Índice: np.argmin(array)
- Desviación estándar: np.std(array)
- Raíz cuadrada: np.sqrt(array)
- Valor absoluto: np.abs(array)

### Filtrado booleano
- Filtrar: array[array > valor]
- Contar: len(array[array > valor])

### Coeficiente de variación
- CV = (np.std(array) / np.mean(array)) * 100
- < 15% → baja variación
- 15-30% → moderada
- > 30% → alta variación

## 🐼 Pandas

### Crear estructuras
- Series: pd.Series([1,2,3], index=["a","b","c"])
- DataFrame: pd.DataFrame({"col1": [...], "col2": [...]})

### Explorar DataFrame
- Filas/columnas: df.shape
- Nombres columnas: df.columns
- Tipos de datos: df.dtypes
- Primera filas: df.head()
- Resumen estadístico: df.describe()

### Acceder a datos
- Una columna: df["columna"] → Series
### Cargar y explorar datos
- Cargar CSV: pd.read_csv("ruta.csv")
- Primeras filas: df.head()
- Info general: df.info()
- Resumen estadístico: df.describe()
- Contar nulos: df.isnull().sum()
- Seleccionar varias columnas: df[["col1", "col2"]]
