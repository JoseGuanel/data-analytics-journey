# Semana 2 - Día 3: Selección de filas con .loc y .iloc

Hoy aprenderé a filtrar y seleccionar filas específicas
del dataset, no solo columnas completas.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../datasets/titanic.csv")

# .iloc - por posición

print(df.iloc[0:3])    # primeras 3 filas
```

```python
print(df.iloc[0:3]) #iloc muestra hasta la ultima iteracion -1
```

```python
print(df.loc[0:3]) #iloc muestra la ultima iteracion
```

```python
# Solo mujeres
mujeres = df.loc[df["Sex"] == "female"]
print(mujeres.shape)

# Solo primera clase
primera_clase = df.loc[df["Pclass"] == 1]
print(primera_clase.shape)
```

## Ejercicio Día 3 - Filtrando el Titanic

Con .loc[] responde:

1. ¿Cuántos hombres había a bordo?
2. ¿Cuántos pasajeros de tercera clase (Pclass=3) había?
3. ¿Cuántos pasajeros eran menores de 18 años?
4. ¿Cuál fue la edad promedio de los pasajeros de primera clase?
5. Combina condiciones: ¿cuántas mujeres había en primera clase?
   (pista: usa & para combinar dos condiciones, y cada condición
   entre paréntesis)

```python
#¿Cuántos hombres había a bordo?
hombres= df.loc[df["Sex"]=="male"]
print("Habian",hombres.shape[0],"hombres a bordo.")
#¿Cuántos pasajeros de tercera clase (Pclass=3) había?
pclass3= df.loc[df["Pclass"] == 3]
print("Habian", pclass3.shape[0],"pasajeros de tercera clase.")
#¿Cuántos pasajeros eran menores de 18 años?
menor= df.loc[df["Age"] < 18]
print("Habian",menor.shape[0],"menores a bordo.")
#¿Cuál fue la edad promedio de los pasajeros de primera clase?
pclass1= df.loc[df["Pclass"] == 1]
print("La edad promedio de los pasajeros de pimera clase es: ",np.mean(pclass1["Age"]))
#Combina condiciones: ¿cuántas mujeres había en primera clase? (pista: usa & para combinar dos condiciones, y cada condición entre paréntesis)

print("Habian",df.loc[(df["Sex"]=="female") & (df["Pclass"]==1)].shape[0],"mujeres de primera clase a bordo")
```
