# Semana 2 - Día 4: Crear y transformar columnas

Hoy aprenderé a crear columnas nuevas a partir de
operaciones con columnas existentes.

```python
import pandas as pd

df = pd.read_csv("../datasets/titanic.csv")

# Crear columna nueva: edad en meses
df["Age_meses"] = df["Age"] * 12

print(df[["Name", "Age", "Age_meses"]].head())
```

```python
# Crear una columna categórica con condición
df["Categoria_edad"] = df["Age"].apply(lambda x: "Menor" if x < 18 else "Adulto")

print(df[["Name", "Age", "Categoria_edad"]].head(10))
```
