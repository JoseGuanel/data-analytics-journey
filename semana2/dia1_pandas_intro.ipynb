# Semana 2 - Día 1: Introducción a Pandas

Pandas es la librería principal para trabajar con
tablas de datos en Python. Hoy aprenderé qué es un
DataFrame y una Series, y la diferencia con NumPy.

```python
import pandas as pd
import numpy as np

# Una Series es como una columna de Excel con etiquetas
ventas = pd.Series([150, 230, 180, 95, 310], 
                     index=["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"])

print(ventas)
print(type(ventas))
```

```python
print(ventas["Lunes"])
print(ventas["Miércoles"])
```

```python
datos = {
    "Vendedor": ["Ana", "Luis", "Carla", "Pedro"],
    "Ventas": [300, 450, 280, 510],
    "Mes": ["Enero", "Enero", "Enero", "Enero"]
}

df = pd.DataFrame(datos)
print(df)
```

```python
print(df.shape)        # filas y columnas
print(df.columns)      # nombres de las columnas
print(df.dtypes)       # tipo de dato de cada columna
print(df["Vendedor"])  # acceder a una columna específica
```

```python
print(type(df["Vendedor"]))
```

## Ejercicio Día 1 - Explorando un DataFrame

Crea un DataFrame con información de 5 productos de una tienda:
- Nombre del producto
- Precio
- Cantidad en stock
- Categoría

Luego responde:
1. ¿Cuántas filas y columnas tiene?
2. ¿Cuáles son los nombres de las columnas?
3. ¿Qué tipo de dato tiene cada columna?
4. Muestra solo la columna de Precio
5. ¿Qué tipo de objeto es esa columna?

```python
#Creo la tabla de datos
datos = {
    "producto":["Bleu de Chanel","Light Blue","Sauvage","Black Opium","Acqua di Gio"],
    "precio":[85000,62000,78000,70000,65000],
    "stock":[10,20,21,5,14],
    "categoria":["Hombre","Mujer","Hombre","Mujer","Hombre"]
}

df= pd.DataFrame(datos)
print(df)
```

```python
# 1. ¿Cuántas filas y columnas tiene?
print(df.shape)
```

```python
# 2. ¿Cuáles son los nombres de las columnas?
print(df.columns)
```

```python
#3. ¿Qué tipo de dato tiene cada columna?
print(df.dtypes)
```

```python
#4. Muestra solo la columna de Precio
print(df["precio"])
```

```python
# 5. ¿Qué tipo de objeto es esa columna? 
print(type(df["precio"]))
```
