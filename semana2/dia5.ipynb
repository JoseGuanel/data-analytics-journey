# Semana 2 - Día 5: Apply y Lambda avanzado

Hoy profundizo en .apply() con funciones más complejas
y aprendo a aplicar funciones a varias columnas a la vez.

```python
import pandas as pd

df = pd.read_csv("../datasets/titanic.csv")

# Función normal en vez de lambda
def categorizar_edad(edad):
    if pd.isna(edad):
        return "Desconocido"
    elif edad < 12:
        return "Niño"
    elif edad < 18:
        return "Adolescente"
    elif edad < 60:
        return "Adulto"
    else:
        return "Adulto Mayor"

df["Grupo_edad"] = df["Age"].apply(categorizar_edad)
print(df[["Name", "Age", "Grupo_edad"]].head(10))
```

```python
def info_pasajero(fila):
    return f"{fila['Name']} - {fila['Sex']} - {fila['Pclass']}era clase"

df["Resumen"] = df.apply(info_pasajero, axis=1)
print(df["Resumen"].head())
```

## Ejercicio Día 5 - Apply avanzado en el Titanic

1. Crea una función "categorizar_tarifa" que clasifique el Fare en:
   - "Económico" si Fare < 10
   - "Medio" si Fare entre 10 y 50
   - "Caro" si Fare > 50
   Aplícala con .apply() y crea la columna "Categoria_tarifa"

2. ¿Cuántos pasajeros pagaron tarifa "Cara"?

3. Crea una función con axis=1 que combine "Pclass" y "Sex" 
   en un texto como: "Mujer de 1era clase" o "Hombre de 3era clase"
   Guárdalo en una columna "Perfil"

4. ¿Cuál fue el grupo (de la columna Perfil) con más pasajeros?
   (pista: investiga la función value_counts())

```python
def categorizar_tarifa(X):
    if (X < 10):
        return "Economico"
    elif (X>=10) & (X<=50):
        return "Medio"
    else:
        return "Caro"

df["Categoria_tarifa"]= df["Fare"].apply(categorizar_tarifa)
df["Categoria_tarifa"]
#¿Cuántos pasajeros pagaron tarifa "Cara"?
print(df.loc[df["Categoria_tarifa"]=="Caro"].shape[0],"Pasajeros pagaron la tarifa Cara")
```

```python
#Crea una función con axis=1 que combine "Pclass" y "Sex" en un texto como: "Mujer de 1era clase" o "Hombre de 3era clase" Guárdalo en una columna "Perfil"
def sex_pclass(fila):
    return f"{fila['Sex']} de {fila['Pclass']}era clase"
df["Perfil"]= df.apply(sex_pclass,axis=1)
df["Perfil"]
```

```python
#¿Cuál fue el grupo (de la columna Perfil) con más pasajeros? (pista: investiga la función value_counts())
print("El grupo con mas pasajeros es: ",df["Perfil"].value_counts().idxmax(),"con",df["Perfil"].value_counts().max(),"pasajeros")
```
