## Ejercicio final integrador
Voy a resolver un problema completo desde cero
usando todo lo aprendido esta semana.

Tienes 5 estudiantes con notas en 4 ramos:

notas = [
    [55, 72, 88, 61],   # estudiante 1
    [90, 85, 78, 92],   # estudiante 2
    [40, 55, 60, 48],   # estudiante 3
    [75, 80, 70, 85],   # estudiante 4
    [65, 58, 72, 69]    # estudiante 5
]

Responde con código:
1. Promedio de cada estudiante
2. Promedio de cada ramo
3. ¿Qué estudiante tiene el mejor promedio?
4. ¿Qué ramo fue el más difícil? (menor promedio)
5. ¿Cuántos estudiantes tienen promedio sobre 70?
6. Gráfico de barras con el promedio de cada estudiante
   - Eje X: Estudiante 1, 2, 3, 4, 5
   - Eje Y: Promedio
   - Título: "Promedio por estudiante"
   - Color: verde

```python
import numpy as np
import matplotlib.pyplot as plt
notas = [
    [55, 72, 88, 61],   # estudiante 1
    [90, 85, 78, 92],   # estudiante 2
    [40, 55, 60, 48],   # estudiante 3
    [75, 80, 70, 85],   # estudiante 4
    [65, 58, 72, 69]    # estudiante 5
]

prom_estudiantes= np.mean(notas, axis=1)
prom_ramo= np.mean(notas, axis=0)
print("Promedios de estudiantes:")
for x in range(len(prom_estudiantes)):
    print("Estudiante ",x+1,": ", prom_estudiantes[x])

print("El estudiante con mejor promedio es el ", np.argmax(prom_estudiantes)+1, "con promedio: ", prom_estudiantes[np.argmax(prom_estudiantes)])

print("El ramo mas dificil es el ", np.argmin(prom_ramo)+1," con promedio general de: ", prom_ramo[np.argmin(prom_ramo)])

print("Hay ", len(prom_estudiantes[prom_estudiantes >70]),"estudiantes con promedio sobre 70")

plt.bar(np.arange(1,len(prom_estudiantes)+1), prom_estudiantes, color="green")
plt.xlabel("Estudiantes")
plt.ylabel("Promedios")
plt.title("Promedio por estudiantes")
plt.show()
```
