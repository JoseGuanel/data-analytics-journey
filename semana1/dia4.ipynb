# Día 4 — Estadísticas con datos reales simulados

Vamos a analizar datos de sueldos, notas y 
temperaturas mensuales usando NumPy.
El objetivo es practicar estadísticas y filtrado
con volúmenes de datos más grandes.

```python
import numpy as np

# Semilla fija — esto hace que todos generemos los mismos números
np.random.seed(42)

# 100 sueldos aleatorios entre $400.000 y $1.500.000
sueldos = np.random.randint(400000, 1500000, size=100)

print("Primeros 10 sueldos:", sueldos[:10])
print("Total de datos:", len(sueldos))
```

## Ejercicio — Análisis de Sueldos

Con el array de 100 sueldos responde con código:

1. ¿Cuál es el sueldo promedio?
2. ¿Cuál es el sueldo máximo y mínimo?
3. ¿Cuántas personas ganan más de $1.000.000?
4. ¿Cuántas personas ganan menos de $500.000?
5. ¿Cuál es la desviación estándar? ¿Es alta o baja?
6. ¿Cuál es el sueldo promedio solo de quienes ganan más de $1.000.000?

```python
print("el sueldo promedio es: ", np.mean(sueldos))
print("el sueldo min es: ", np.min(sueldos)," y el sueldo max es: ", np.max(sueldos))
print(len(sueldos[sueldos>1000000])," ganan mas de 1.000.000 y ", len(sueldos[sueldos<500000])," ganan menos de 500.000")
print("la desviacion estandar es: ", np.std(sueldos))
print("el sueldo promedio de quienes ganan mas de 1m es: ", np.mean(sueldos[sueldos>1000000]))

print("cv: ",(np.std(sueldos) / np.mean(sueldos) ) * 100)
```
