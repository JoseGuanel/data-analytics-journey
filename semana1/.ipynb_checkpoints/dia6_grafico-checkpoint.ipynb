# Día 6 — Mini Proyecto: Análisis de Temperaturas

Analizaré las temperaturas de una ciudad durante 30 días.
Calcularé estadísticas clave y visualizaré los datos
en un gráfico para identificar patrones.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(7)
temperaturas = np.random.randint(15, 35, size=30)
dias = np.arange(1, 31)

print("Temperaturas:", temperaturas)
```

## Análisis completo de temperaturas — 30 días

Con el array de temperaturas responde con código:

1. Temperatura promedio, máxima y mínima del mes
2. ¿Cuántos días superaron los 28 grados?
3. ¿Cuál fue el día más caluroso? (número de día, no índice)
4. Desviación estándar e interpretación con CV
5. Graficar las temperaturas de los 30 días con matplotlib
   - Eje X: días (1 al 30)
   - Eje Y: temperatura
   - Título: "Temperaturas del mes"
   - Color de línea: rojo

```python
import matplotlib.pyplot as plt
print("Temperatura promedio: ", np.mean(temperaturas),"°")
print("Temperatura maxima: ", np.max(temperaturas),"° | Temperatura minima: ", np.min(temperaturas),"°")
print("El dia mas caluroso fue el dia: ", dias[np.argmax(temperaturas)], "con ", np.max(temperaturas),"°") 
print("Desviacion estandar: ", np.std(temperaturas), "cv: ",(np.std(temperaturas) / np.mean(temperaturas))*100)

plt.plot(dias, temperaturas, color="red")
plt.title("Temperaturas del mes")
plt.xlabel("Días")
plt.ylabel("Temperatura °C")
plt.show()
```
