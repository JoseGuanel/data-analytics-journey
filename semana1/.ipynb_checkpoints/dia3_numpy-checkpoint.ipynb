```python
import numpy as np

# Datos de ejemplo — imagina que son temperaturas de una semana
temperaturas = np.array([18, 22, 25, 19, 30, 27, 21])

# Estadísticas básicas
print("Promedio:", np.mean(temperaturas))
print("Máxima:", np.max(temperaturas))
print("Mínima:", np.min(temperaturas))
print("Total:", np.sum(temperaturas))
print("Desviación estándar:", np.std(temperaturas))
```

```python
# Funciones matemáticas útiles
print("Raíz cuadrada:", np.sqrt(temperaturas))
print("Redondear:", np.round(temperaturas * 1.8 + 32, 1))  # convertir a Fahrenheit
print("Valor absoluto:", np.abs(np.array([-3, 5, -1, 8, -2])))

# Comparaciones — esto es MUY importante para filtrar datos
print("\nDías con más de 22 grados:", temperaturas[temperaturas > 22])
print("Días con menos de 20 grados:", temperaturas[temperaturas < 20])
```

Día 3 — Ejercicio Final: Análisis de Ventas

Tengo los datos de ventas de una tienda durante 7 días:
ventas = [150, 230, 180, 95, 310, 275, 200]

Responde lo siguiente con código NumPy:

1. ¿Cuál fue el valor máximo y mínimo de ventas?
2. ¿Cuál es el promedio de ventas de la semana?
3. ¿Cuántos días superaron las 200 ventas?
4. ¿Cuál fue la desviación estándar? ¿Los datos son muy dispersos o no?
5. Convierte todas las ventas a dólares (multiplica por 0.0011)

```python
ventas = np.array([150, 230, 180, 95, 310, 275, 200])
print("maximo: ", np.max(ventas))
print("minimo: ", np.min(ventas))
print("promedio: ", np.mean(ventas))
print(">200: ", len(ventas[ventas>200]))
print("desviacion estandar: ", np.std(ventas))
print("ventas a dolares: ", ventas*0.0011)
```
